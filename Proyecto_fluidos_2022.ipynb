## Introducción

<p align="justify">En este informe se estudiará el fenómeno de turbulencia con el objetivo de observar el comportamiento de un fluido para llevarlo a un análisis matemático, esto mediante el uso de los conocimientos adquiridos en el curso y conceptos pertinentes al curso, esto con la finalidad de poder describir la naturaleza y propiedades físicas del fenómeno que estudiaremos.

<p align="justify">El estudio de este tipo de fenómenos tiene un gran impacto en la sociedad actual, ya que nos permite diseñar y evaluar equipos y procesos en campos tecnológicos tan diversos como el transporte de fluidos, generación de energía, control ambiental, estructuras hidráulicas, entre otros.

<p align="justify">La turbulencia es un fenómeno de la naturaleza que se caracteriza por tener flujos caóticos, irregulares y poco predecibles, además de poseer múltiples vórtices o remolinos. Este fenómeno es causado por la predominancia de las fuerzas inerciales por sobre las viscosas, lo que es equivalente a un alto número de Reynolds.

<p align="justify">Debido al alto grado de desorden que presenta este fenómeno su estudio se dificulta, ya que depende de variables que son difíciles de estudiar entre las cuales podemos destacar la vorticidad y la compresibilidad.


## Marco Teórico

<p align="justify">Cuando estudiamos el flujo de los fluidos en la naturaleza, estos son en su mayoría turbulentos. El parámetro fundamental que nos permite diferenciar flujos laminares de turbulentos es el número de Reynolds, que representa una relación entre las fuerzas inerciales y viscosas del fluido.

<p align="justify">Un fluido turbulento presenta las siguientes características.

<ul>
  <li>Altos números de Reynolds</li>
  <li>Flujo irregular, impredecible y caótico</li>
  <li>Problema no-lineal</li>
  <li>Vorticidad: Presenta remolinos que interactuan entre ellos causando disipación de energía</li>
  <li>Difusividad: Altos niveles de calor, transporte de masa y de cantidad de movimiento</li>
</ul>




<p align="justify">Debido a la dificultad y complejidad de los flujos turbulentos es necesario hacer un análisis estadístico. Por ende ahora procederemos a explicar algunos conceptos que se usarán en el futuro.

**Media:**

<p align="justify">Este concepto representado como $<u>$ es el valor promedio de un conjunto de datos. Usando la siguiente fórmula para obtenerlo cuando el conjunto de datos es discretos.

<p align="center">$<u> = \frac{1}{N} ∑_{i=1}^{N} u_{i}$

**Varianza**

<p align="justify">La varianza representa la dispersión de los datos con respecto a la media y su respectiva fórmula es la siguiente para un conjunto de datos discretos.

<p align="center">$σ^{2} = \frac{1}{N-1} ∑_{i=1}^{N} (u_{i} - <u>)^{2}$

**Covarianza Cruzada**

<p align="justify">Es una medida de similitud entre dos conjuntos de datos, es usada para comparar la relación entre estos dos conjuntos. Además, la covarianza puede presentar un desfase $τ$ en la comparación de los conjuntos.

<p align="center">$cov[u(t), v(t+τ)] = \frac{1}{N-1} ∑_{i=1}^{N} (u_{i} - <u>)(v_{i+τ}-<v>)$

**Correlación Cruzada**

<p align="justify">Falta detallar

<p align="center">$ρ_{uv}(τ) = \frac{cov[u(t), v(t+τ)]}{σ_{u}σ_{v}}$

**Función de Autocorrelación**

<p align="justify">Esta función nos ayuda a ser capaz de evaluar la memoria de una señal en donde vemos como los valores previos afectan a los futuros.

<p align="center">$ρ_{uv}(τ) = \frac{cov[u(t), v(t+τ)]}{σ_{u}σ_{v}}$

**Descomposición de Reynolds**

<p align="justify">Las velocidades en un flujo turbulento se pueden representar como

<p align="center">$u = <U> + u’$

<p align="justify">En donde $u’$ es la diferencia entre la velocidad instantánea y la velocidad media. Como muestra en la siguiente figura.

<p align="justify">En donde $<u’>$ es cero.

**Cascada de Energía**

<p align="justify">El fenómeno de la turbulencia tiene sub-categorías. Entre estas sub-categorías encontramos la turbulencia isotrópica, la cual está definida como un tipo de turbulencia donde el valor de cualquier función de sus componentes de velocidad y derivadas en un punto, definido en relación a un conjunto particular de ejes, permanecen inalterados si los ejes de referencia son rotados en cualquier dirección, o si el sistema coordenado es reflejado en cualquier plano a través del origen, es decir, presenta simetrías espaciales entorno al origen.

<p align="justify">El efecto principal de la turbulencia es que disipa la energía del sistema. Esto mediante la generación de vórtices, que al principio tienen un mayor diámetro, dado que poseen más energía y menor frecuencia. Con el transcurso del tiempo disminuyen su diámetro, este efecto se genera debido a la inestabilidad de los vórtices, ya que solo duran un tiempo limitado y posteriormente van chocando y haciéndose más pequeños a través de la transferencia de energía, desde los de mayor tamaño a los de menor diámetro, lo que se conoce como cascada de energía, hasta que su tamaño es tan pequeño que la viscosidad es capaz de disipar su energía en forma de calor. Antes de la disipación por calor y después de las grandes escalas, que es cuando los vórtices son muy grandes, existe un periodo intermedio denominado subrango inercial, y en este, la tasa de transferencia de energía es constante e igual a -5/3. El choque de los vórtices provoca que se intercambien elementos de fluido entre las diferentes capas que lo componen.

**Esfuerzo de Reynolds**
<p align="justify">En un flujo turbulento unidireccional.

<p align="center">$v = (U + u’) \widehat{i} + v’\widehat{j} + w’\widehat{k}$


WhatsApp Image 2022-11-27 at 11.15.09 PM.jpeg

<p align="justify">En la cantidad de movimiento inducido por la fluctuación vertical en el eje x será de: 

<p align="center">$dF = \rho v’\triangle x \triangle z u’$

<p align="justify">Por lo que la fuerza instantánea en la cara superior del elemento diferencial es:

<p align="center">$dF = \rho v’ u’ dA$

<p align="justify">Estas fluctuaciones inducen una fuerza que es análoga a los esfuerzos viscosos sobre el flujo. Este efecto se cuantifica usando una matriz, la cual se llama tensor de esfuerzos de Reynolds:

<p align="justify">La diagonal del tensor es la energía cinética turbulenta que cuantifica la cantidad de fluctuación:

<p align="center">$k = \frac{1}{2} (<u’^{2}><v’^{2}>)$

###Librerías utilizadas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

###Extracción de datos de archivos





```python
u=np.genfromtxt('u.csv', delimiter=',')
v=np.genfromtxt('v.csv', delimiter=',')
x=np.genfromtxt('x.csv', delimiter=',')
y=np.genfromtxt('y.csv', delimiter=',')
# ajustar tamanos
u=np.transpose(u.reshape(127,2999,29), (0, 2, 1))
v=np.transpose(v.reshape(127,2999,29), (0, 2, 1))
x=np.transpose(x.reshape(127,29), (0, 1))
y=np.transpose(y.reshape(127,29), (0, 1))
##### aplicar constantes ¿?  <----
c = 1         # en mm.
t = 1/1000     # tiempo entre matrices en seg.

u = u    # velocidades en u en mm/seg.
v = v # velocidades en v en mm/seg.
x = x*c       # posiciones en x en mm.
y = y*c    
   # posiciones en y en mm.
```

### Velocidad Media

```python
#Velocidad promedio del espacio entero en un instante
u_promedio=u.mean(axis=(0,1))
u_prima=u_promedio-np.mean(u_promedio)
v_promedio=v.mean(axis=(0,1))
v_prima=v_promedio-np.mean(v_promedio)
velocidad_promedio=np.linalg.norm((u_promedio, v_promedio), axis=0)

f=velocidad_promedio
n=len(f)
dt=t
t_=dt*np.arange(n)

fig,axis=plt.subplots(3,1)

plt.sca(axis[0])

plt.plot(t_,u_promedio,color='c')
plt.axhline(y=np.mean(u_promedio), color='r', linestyle='-')
plt.xlabel('Tiempo [s]')
plt.ylabel('Velocidad u [mm/s] ')

plt.sca(axis[1])
plt.plot(t_,v_promedio,color='c')
plt.axhline(y=np.mean(v_promedio), color='r', linestyle='-')
plt.xlabel('Tiempo [s]')
plt.ylabel('Velocidad v [mm/s]')

plt.sca(axis[2])
plt.plot(t_,velocidad_promedio,color='c')
plt.axhline(y=np.mean(velocidad_promedio), color='r', linestyle='-')
plt.xlabel('Tiempo [s]')
plt.ylabel('Velocidad  [mm/s]')


```

```python
plt.plot(t_,u_promedio,color='c', label='u(t)')
plt.axhline(y=np.mean(u_promedio), color='r', linestyle='--', label='<u>')
randnums= np.random.randint(1,100, size=len(u_promedio))
plt.xlabel('Tiempo [s]')
plt.ylabel('Velocidad u [mm/s] ')
plt.legend()
```

```python
plt.plot(t_,velocidad_promedio,color='c')
plt.axhline(y=np.mean(velocidad_promedio), color='r', linestyle='-')
plt.xlabel('Tiempo [s]')
plt.ylabel('Velocidad  [mm/s]')
```

###Varianza

```python
varianza_v=np.var(v_promedio)
varianza_u=np.var(u_promedio)
varianza_velocidad=np.var(velocidad_promedio)

print(f'La varianza de la velocidad en todos los puntos del espacio es {varianza_velocidad}')
```

###Autocorrelación


#### Autocorrelación de la velocidad ${u}$

```python
autocorrelacion_u = np.correlate(u_promedio, u_promedio, mode = 'full')
autocorrelacion_u = autocorrelacion_u[autocorrelacion_u.size // 2 :]
x = pd.plotting.autocorrelation_plot(autocorrelacion_u) 
x.plot() 
plt.show()
```

La escala integral de tiempo de velocidad ${u}$

```python
escala_integral_tiempo_u=sum(autocorrelacion_u)/varianza_u
print(escala_integral_tiempo_u)
```

####Autocorrelación de la velocidad ${v}$

```python
autocorrelacion_v = np.correlate(v_promedio, v_promedio, mode = 'full')
autocorrelacion_v = autocorrelacion_v[autocorrelacion_v.size // 2 :]
y = pd.plotting.autocorrelation_plot(autocorrelacion_v) 
y.plot() 
plt.show()
```

La escala integral de tiempo de velocidad ${v}$

```python
escala_integral_tiempo_v=sum(autocorrelacion_v)/varianza_v
print(escala_integral_tiempo_v)
```

###Correlación cruzada, entre la velocidad ${u}$ y ${v}$

```python
u_punto_promedio=u.mean(axis=(2))
v_punto_promedio=v.mean(axis=(2))
flactuacion_u=[]
flactuacion_v=[]
u_cambio=np.transpose(u.reshape(127,2999,29), (1, 0, 2))
v_cambio=np.transpose(v.reshape(127,2999,29), (1, 0, 2))
for u_ in u_cambio:
  flactuacion_u.append(u_-u_punto_promedio)
for v_ in v_cambio:
  flactuacion_v.append(v_-v_punto_promedio)
flactuacion_v=np.array(flactuacion_v)
flactuacion_u=np.array(flactuacion_u)



plt.scatter(flactuacion_u.reshape(-1),flactuacion_v.reshape(-1),2)
plt.xlabel("Velocidad u' [mm/s]")
plt.ylabel("Velocidad v' [mm/s]")
plt.grid()
plt.savefig('comparacion_velocidades')
```

```python
correlacion_cruzada = np.correlate(u_promedio, v_promedio, mode = 'full')
correlacion_cruzada = correlacion_cruzada[correlacion_cruzada.size // 2 :]
print(correlacion_cruzada[200]/(varianza_u*varianza_v))
y = pd.plotting.autocorrelation_plot(correlacion_cruzada) 
y.plot() 
plt.ylabel('correlacion cruzada')
plt.show()
```

```python
covarianza=np.cov(u_promedio, v_promedio)
correlacion_cruzada=(covarianza/(np.sqrt(varianza_u*varianza_v)))[0][1]
print(correlacion_cruzada)
```

###Distribución de esfuerzos de Reynolds

```python
tensor_reynolds=np.cov(u_promedio, v_promedio)
print(tensor_reynolds)
```

###Energía cinética turbulenta (TKE)

```python
TKE=0.5*(varianza_u+varianza_v)
print(f'La distribución espacial de energía cinética turbulenta es de {TKE}')
```

###Densidad espectral de energía (PSD)

```python
f=velocidad_promedio
n=len(f)
dt=t
t_=dt*np.arange(n)


#FFT y DFT
fhdt=np.fft.fft(f,n)
psd=fhdt*np.conj(fhdt)/n
freq=(1/dt)*np.arange(n)
l=np.arange(0,np.floor(n/2),dtype='int')

filtro=psd>750
fhdt_clean=fhdt*filtro
f_clean=np.fft.ifft(fhdt_clean)

fig,axis=plt.subplots(2,1)

plt.sca(axis[0])
plt.plot(t_,f,color='c',label='ruidosa')
plt.plot(t_,f_clean,color='k',label='limpia')
plt.xlabel('Tiempo [s]')
plt.ylabel('Velocidad [mm/s]')
plt.legend()
print(l)
plt.sca(axis[1])
plt.plot(freq[l],psd[l],color='c',)
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Frecuencia [Hz]')
plt.xscale('log')
plt.yscale('log')

```
